# PROCESSAMENTO

###1. Consolidar os pagamentos de férias, identificando nas planilhas mensais de férias para pagamento da SPP, os pagamentos de férias realizados, e relacionando-os com as informações correspondentes planilhas mensais de relação por banco da SRF, de forma a vincular cada pagamento ao respectivo período aquisitivo.
###2. Nos casos em que não for localizada correspondência do pagamento e seu período aquisitivo, registrar essa ausência de vínculo de forma destacada.


```python
# GERA O DATAFRAME DE SAÍDA, QUE SERÁ A CONTAGEM DE VEZES QUE CADA SERVIDOR APARECE NAS PLANILHAS DE PAGAMENTO FÉRIAS
colunas_resumo = ['MATRÍCULA', 'NOME_SERVIDOR', 'QTD_OCORRÊNCIA', 'ARQUIVO_01', 'PER_AQUISITIVO_01', 'VALOR_01'] #, 'ARQUIVO02', 'ARQUIVO03', 'ARQUIVO04','ARQUIVO05']

out_resumo_df = pd.DataFrame(columns=colunas_resumo)


#CONFERÊNCIA DAS OCORRÊNCIAS

#Conferir nas planilhas de relação por banco da SPP mês a mês, todas as ocorrências de
#pagamentos de férias para cada servidor. Para cada planilha mensal da SPP, procurar a correspondente planilha de Férias para pagamento da SRF,
#procurando as correspondências dos servidores nas duas planilhas, e obtendo o período aquisitivo de cada pagamento

#Gerar um dataframe resumo, com todos os valores de ⅓ de férias recebidos por servidor (obtidos na planilha da SPP), com os correspondentes
#períodos aquisitivos (obtidos na planilha da SRF) . Caso não seja encontrada a ocorrência de férias na planilha de férias para pagamento da SRF,
#marcar a coluna “Período Aquisitivo” com um símbolo de não encontrado


#itera nas planilhas na pasta
for planilha in sorted(path_planilhas_ferias_spp.iterdir()): #itera nos arquivos da pasta
  if planilha.is_dir(): # Só continua a iteração se for um arquivo
    continue #encerra esta iteração

  print(planilha.name)
  #print()

  # Obtém o numero do arquivo (2 primeiros caracteres do nome)
  ordem_arquivo = planilha.name[0:2]
  #print(ordem_arquivo)

  #obtém a planilha da SRF correspondente a planilha (da SPP)
  #print(ordem_arquivo + '*.*')
  #planilha_srf = ''
  #for plan_srf in path_planilhas_ferias_srf.rglob(ordem_arquivo + '*.*'):
  for plan_srf in path_planilhas_ferias_srf.glob(ordem_arquivo + '*.*'): # glob não entra em subpastas
    print(plan_srf.name)
    planilha_srf = (path_planilhas_ferias_srf  / plan_srf.name )

  #print(planilha_srf)

  #break



  #gera um DF correspondente à planilha da iteração, e da planilha da SRF correspondente
  in_ferias_df = pd.read_excel(path_planilhas_ferias_spp  / planilha.name )
  #display(in_ferias_df)

  #gera um DF da planilha da SRF
  plan_SRF_DF = pd.read_excel(planilha_srf)
  #display(plan_SRF_DF)
  #break

  #percorre o df para fazer a verificação
  for linha in range(len(in_ferias_df)):
    matricula_in = in_ferias_df.loc[linha,'MATRÍCULA']
    #print(matricula_in)

    # Obtém o período aquisitivo daquele servidor, obtido da planilha da SRF
    qtd_srf = len(plan_SRF_DF.loc[plan_SRF_DF['Matrícula'] == matricula_in]) # essa qtd vai ser 0 ou 1
    if qtd_srf > 0: #encontrou o servidor na planilha da SRF
      periodo_aquisitivo = str(int(plan_SRF_DF.loc[plan_SRF_DF['Matrícula'] == matricula_in, 'Per. Aquisit.']))
      #display(plan_SRF_DF.loc[plan_SRF_DF['Matrícula'] == matricula_in])
      #break
    else: # Não encontrou
      periodo_aquisitivo = 'NÃO ENCONTRADO'

    #print(periodo_aquisitivo)
    ###########
    #break

    #para a linha da interação, verifica se o servidor já está registrado no DF de saída (out_resumo_df)
    qtd = len(out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in]) # essa qtd vai ser 0 ou 1
    if qtd > 0: #já existe linha para o servidor (nesse caso, será 1 linha)
      # se já existe, então incrementa o número de ocorrências do servidor e acrescenta a coluna com o nome do arquivo da planilha
      out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'QTD_OCORRÊNCIA'] +=1
      out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'ARQUIVO_0'+str(int(out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'QTD_OCORRÊNCIA']))] = planilha.name

      #Atribui também o período aquisitivo e o valor recebido líquido
      out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'PER_AQUISITIVO_0'+str(int(out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'QTD_OCORRÊNCIA']))] = periodo_aquisitivo
      out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'VALOR_0'+str(int(out_resumo_df.loc[out_resumo_df['MATRÍCULA'] == matricula_in, 'QTD_OCORRÊNCIA']))] = in_ferias_df.loc[linha,'VALOR']

    else: # nesse caso, não existe a ocorrência desse servidor no DF de saída (out_resumo_df)
      #então insere a linha correspondente ao servidor, com QTD_OCORRÊNCIA setado para 1
      nova_linha = {
          'MATRÍCULA': [in_ferias_df.loc[linha,'MATRÍCULA']],
          'NOME_SERVIDOR': [in_ferias_df.loc[linha, 'NOME SERVIDOR']],
          'QTD_OCORRÊNCIA': [1],
          'ARQUIVO_01': [planilha.name],
          'PER_AQUISITIVO_01': [periodo_aquisitivo], #obtido da planilha da SRF
          'VALOR_01': [in_ferias_df.loc[linha,'VALOR']]
      }
      #print(nova_linha)

      out_resumo_df = pd.concat([out_resumo_df, pd.DataFrame(nova_linha)]) #, ignore_index=True)

#ordena descendente por qtd_ocorrências e  ascendente por nome_servidor
out_resumo_df = out_resumo_df.sort_values(['QTD_OCORRÊNCIA', 'NOME_SERVIDOR'] , ascending = [False, True], ignore_index=True)


display(out_resumo_df)
```

###3. Localizar na planilha de rubricas de folha de pagamento (planilhão), eventuais rubricas correspondentes à devoluções de valores referentes ao ⅓ de férias, vinculando-as aos servidores e às respectivas folhas de pagamento.


```python
#Verificar se os servidores devolveram valores referentes a pagamentos de
# ⅓ de férias, e acrescentar eventuais devoluções no dataframe resumo

# Lê o planilhão com as rubricas
rubricas_df = pd.read_excel(path_planilhao_rubricas)

#Filtrar na planilha de rubricas (planilhão), apenas as linhas referentes à rubricas
# de devoluções de valores recebidos à título de ⅓ de férias (“Cod Rubrica” == 8041000),
# gerando um dataframe com as devoluções (linhas filtradas)

# seleciona só as linhas com rubrica de devolução de 1/3 de férias
devolucoes_ferias_df = rubricas_df.loc[rubricas_df['Cod Rubrica'] == 8041000 ]

display(devolucoes_ferias_df)

# Gera planilha Excel com os dados de devoluções de 1/3 de férias
#devolucoes_ferias_df.to_excel(path_rubricas_devolucao_ferias, index = False, header = True)
```

###4. Gerar relatório consolidado: produzir uma planilha final contendo, por servidor, as ocorrências de pagamentos de ⅓ de férias, os períodos aquisitivos correspondentes, as devoluções ocorridas e o número das respectivas folhas de pagamento, bem como demais informações necessárias para análise. Ordenar a planilha pelo número de ocorrências


```python
# No dataframe resumo, acrescentar, linha a linha, os valores devolvidos pelo servidor e a respectiva folha de pagamento (mês/ano) onde ocorreu a devolução.
# EssaS informações serão obtidas no dataframe de devoluções, procurando a correspondência pela matrícula do servidor

#percorre o out_resumo_df para fazer a verificação e registro dos valores de 1/3 de férias devolvidos pelos servidores
for linha_out in range(len(out_resumo_df)):
  matricula_out = out_resumo_df.loc[linha_out,'MATRÍCULA']
  #print(matricula)

  ### para cada servidor do laço, verifica as ocorrêcias dele no rubricas_df (que contem os registros de devoluções de 1/3 de férias) ###
  # DF com as ocorrências de devoluções encontradas para o servidor
  devolucoes_ferias_servidor_df = devolucoes_ferias_df[devolucoes_ferias_df['Matrícula'] == matricula_out].reset_index()
  #display(devolucoes_ferias_servidor_df)


  # percorre devolucoes_ferias_servidor_df e acrescenta os valores no DF resumo de saída (out_resumo_df)
  for linha_devolucao in range(len(devolucoes_ferias_servidor_df)):
    #print(linha_devolucao)
    out_resumo_df.loc[linha_out,'VALOR_DEVOLUCAO_0' + str(linha_devolucao+1)] = devolucoes_ferias_servidor_df.loc[linha_devolucao,'Valor']
    out_resumo_df.loc[linha_out,'FOLHA_DEVOLUCAO_0' + str(linha_devolucao+1)] = str(int(devolucoes_ferias_servidor_df.loc[linha_devolucao,'Mês/Ano Folha']))

  ########################################################################################################################################

  #break

display(out_resumo_df)
```
